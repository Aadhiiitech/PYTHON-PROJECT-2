```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv")
data.head()
```

```python
age_series = data["Age"]
fare_series = data["Fare"]
survived_series = data["Survived"]

print(age_series.head())
```

```python
data.head()
data.info()
data.describe()
```

```python

data["Age"].fillna(data["Age"].mean(), inplace=True)
data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)
```

```python
data.loc[5]
data.loc[0:5, ["Age", "Fare"]]
```

```python
data.rename(columns={"Sex": "Gender"}, inplace=True)
pivot_table = data.pivot_table(values="Survived",
                             index="Gender",
                             columns="Pclass",
                             aggfunc="mean")
```

```python
age_above_30 = data[data["Age"] > 30]
fare_above_50 = data[data["Fare"] > 50]
female_survivors = data[(data["Gender"] == "female") & (data["Survived"] == 1)]

data.sort_values(by="Fare", ascending=False)
data.sort_values(by="Age")
data.sort_values(by="Survived", ascending=False)
```

```python
data.groupby("Gender")["Survived"].mean()
data.groupby("Pclass")["Age"].mean()
data.groupby("Embarked")["Survived"].sum()
```

```python
class_info = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Category": ["Upper", "Middle", "Lower"]
})

merged_data = pd.merge(data, class_info, on="Pclass", how="left")

joined_data = data.join(class_info.set_index("Pclass"), on="Pclass")

concat_data = pd.concat([data.head(), data.tail()])
```

```python
data.groupby("Survived")["Age"].mean()

highest_fare = data["Fare"].max()
lowest_fare = data["Fare"].min()

survival_percentage = data["Survived"].mean() * 100
```
